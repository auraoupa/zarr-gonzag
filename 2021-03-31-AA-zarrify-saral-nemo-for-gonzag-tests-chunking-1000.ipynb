```python
import xarray as xr
import zarr
import numcodecs
```

```python
ds_alt1=xr.open_dataset('/Users/auraoupa/Work/git/gonzag/dt_global_alg_sla_vxxc_20170402_SARAL-Altika.nc',chunks={'time':1000})
```

```python
ds_alt1
```

```python
compressor = numcodecs.Blosc(cname='snappy', clevel=6, shuffle=-1)
encoding = {vname: {'compressor': compressor} for vname in ds_alt1.variables}
outdir = '/Users/auraoupa/Work/git/gonzag/zarr_dt_global_alg_sla_vxxc_20170402_SARAL-Altika_chunksby1000'
ds_alt1.to_zarr(outdir, encoding=encoding, mode="w")
```
